```python
from keras.models import load_model
from mod_gendata import *
import matplotlib.pyplot as plt
```

```python
autoencoder = load_model("./sflp_autoencoder50.h5")
```

```python
# load test dataset
testdir= '/home/kurihana/ml_model/work_mymodel/ex4/data/test_data'
gd     = gen_grads_data()
x_test = gd.load_data(testdir)
```

```python
decoded_imgs = autoencoder.predict(x_test)
```

```python
# settings
n   = 5   # number of pics
lat = 181 # number of latitude 
lon = 360 # number of longitude

# plot
plt.figure(figsize=(16,8))
for i in range(n):
      ax = plt.subplot(2,n,i+1)
      plt.imshow(x_test[i].reshape(lat,lon))
      plt.gray()
      ax.get_xaxis().set_visible(False)
      ax.get_yaxis().set_visible(False)
 
      ax = plt.subplot(2,n,i+1+n)
      plt.imshow(decoded_imgs[i].reshape(lat,lon))
      plt.gray()
      ax.get_xaxis().set_visible(False)
      ax.get_yaxis().set_visible(False)
 
plt.show()
```
